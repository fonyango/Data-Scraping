### 1.0 Importing libraries

```python
import pandas as pd
from pandas.io.html import read_html
```

### 1.1 Extracting all tables on the page

```python
#extract all tables on the page
url_page = 'https://en.wikipedia.org/wiki/List_of_universities_in_South_Africa'
tables = read_html(url_page, attrs={'class':'wikitable sortable'})

#Print the number of tables extracted from the page
print('Extracted {num} tables'.format(num=len(tables)))
```

### 1.2 Concatenate all tables to form one dataframe

```python
all_tables = pd.concat([tables[0],tables[1],tables[2],tables[3],tables[4],tables[5]], axis=0)
all_tables.head()
```

### 1.3 Save as csv file

```python
all_tables.to_csv('Universities.csv', index=False)
```
